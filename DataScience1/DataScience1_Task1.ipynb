First we have to import the library pandas for handling with dataframes and the library matplotlib.pylot for the visualization of our data.

```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import array
```

Next, we define our column headers and read the data into a data frame by specifying the file, passing the column headers and designating the separator. And we print so statistics of th data.

```python
columns = ['Sepal-length', 'Sepal-width', 'Petal-length', 'Petal-width', 'Class']
df = pd.read_table('data.all',names= columns, sep='\s+', header = None)
print(df.describe())
```

The scatter matrix is now carried out here. First we create an array for the colors of the three classification with the colors red, green and blue. Than we call the plotting for the scatter matrix. Here we give the data in form of the dataframe, the alpha value (which represents the amount of transparency applied), the size of the plotting boxes and our colors.

```python
colors=array(50*['r']+50*['g']+50*['b'])
pd.plotting.scatter_matrix(df, alpha=0.2, figsize=(10,10), color=colors)
plt.show()
```

For the boxplot, we only indicate what kind of plot we would like, in this case 'box'.

```python
df.plot(kind='box')
plt.show()
```

From the plots we can now determine in which value ranges the respective Sepal and Petal points are located and thus classify them.
